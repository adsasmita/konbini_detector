```python
import os
import glob
import pandas as pd
import xml.etree.ElementTree as ET

def xml_to_csv(path):
    xml_list = []
    for xml_file in glob.glob(path + '/*.xml'):
        tree = ET.parse(xml_file)
        root = tree.getroot()
        for member in root.findall('object'):
            value = (root.find('filename').text,
                     int(root.find('size')[0].text),
                     int(root.find('size')[1].text),
                     member[0].text,
                     int(member[4][0].text),
                     int(member[4][1].text),
                     int(member[4][2].text),
                     int(member[4][3].text)
                     )
            xml_list.append(value)
    column_name = ['filename', 'width', 'height', 'class', 'xmin', 'ymin', 'xmax', 'ymax']
    xml_df = pd.DataFrame(xml_list, columns=column_name)
    return xml_df
```

```python
# Input: directory='train' or 'test', target_name = 'ngb' or 'ndvibw'
def rename_xml_namefiles(directory='train', target_name='ngb'):
    xml_df = xml_to_csv(os.path.join(os.getcwd(),'images/{}'.format(directory)))

    for i in range(len(xml_df)):
        if xml_df.filename[i].startswith('test_'):
            xml_df.filename[i]=target_name+xml_df.filename[i][4:]
    
    # some XML files in ngb has referred to "ngb_xx_yy" instead of "test_xx_yy"
    if target_name=='ndvibw':
        for j in range(len(xml_df)):
            if xml_df.filename[j].startswith('ngb_'):
                xml_df.filename[j]=target_name+xml_df.filename[j][3:]       
            
    return xml_df
```

```python
# Input: directory='train' or 'test', target_name = 'ngb' or 'ndvibw' or 'ndvigr'
def rename_xml_namefiles(directory='train', target_name='ngb'):
    xml_df = xml_to_csv(os.path.join(os.getcwd(),'images/{}'.format(directory)))
    for i in range(len(xml_df)):
        if xml_df.filename[i].startswith('test_'):
            xml_df.filename[i]=target_name+xml_df.filename[i][4:]
            
    return xml_df
```

## Converting NGB XML Filenames

```python
xml_df_ngb_train = rename_xml_namefiles(directory='train', target_name='ngb')
xml_df_ngb_train
```

```python
xml_df_ngb_test = rename_xml_namefiles(directory='test', target_name='ngb')
xml_df_ngb_test
```

## Converting NDVIBW XML Filenames

```python
xml_df_ndvibw_train = rename_xml_namefiles(directory='train', target_name='ndvibw')
xml_df_ndvibw_train
```

```python
xml_df_ndvibw_test = rename_xml_namefiles(directory='test', target_name='ndvibw')
xml_df_ndvibw_test
```

## Merging Dataframes for NGB and NDVIBW Respectively

```python
xml_df_train = pd.concat([xml_df_ngb_train, xml_df_ndvibw_train], ignore_index=True)
xml_df_train
```

```python
xml_df_test = pd.concat([xml_df_ngb_test, xml_df_ndvibw_test], ignore_index=True)
xml_df_test
```

## Verify Counts of `ngb_` and `ndvibw_`

```python
def count_name(xml_df, starts_with='ndvibw_'):
    count = 0
    for i in range(len(xml_df)):
        if xml_df.filename[i].startswith('ndvibw_'):
            count+=1
    print("Count of rows that starts with `{}`: {}".format(starts_with, count))
```

```python
count_name(xml_df_train, starts_with='ngb_')
count_name(xml_df_train, starts_with='ndvibw_')
count_name(xml_df_test, starts_with='ngb_')
count_name(xml_df_test, starts_with='ndvibw_')
```

## Export CSVs to train and test

```python
xml_df_train.to_csv('csv/train_labels.csv',index=None)
xml_df_test.to_csv('csv/test_labels.csv',index=None)
```
