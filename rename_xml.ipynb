```python
import shutil
import os
from pathlib import Path
```

## Check how many files ends with .png (expect 400, then cut down to 256)

```python
dir = '.'
allfiles = os.listdir(".")
files = [ fname for fname in allfiles if fname.endswith('.png')]
print(len(files))
```

## If name.png is not found in name.xml, delete the file

```python
for file in files:
    xml_fname = Path(file[:-3]+'xml')
    if not xml_fname.is_file():
        os.remove(file)
```
